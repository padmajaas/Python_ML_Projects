```python
cv_train=cv_all[cv_all['data']=='train']
del cv_train['data']
cv_test=cv_all[cv_all['data']=='test']
cv_test.drop(['V86','data'],axis=1,inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn import metrics
```

```python
from sklearn.model_selection import train_test_split
```

```python
params={'class_weight':['balanced',None],
        'penalty':['l1','l2'],
        'C':np.linspace(0.01,1000,10)}
model=LogisticRegression(fit_intercept=True)
```

```python
grid_search=GridSearchCV(model,param_grid=params,cv=5,scoring="roc_auc")
```

```python
x_train=cv_train.drop(['V86','ID'],axis=1)
y_train=cv_train['V86']
x_test=cv_test.drop('ID',axis=1)
cv_train1,cv_train2=train_test_split(cv_train,test_size=0.2,random_state=2)
x_train1=cv_train1.drop(['V86','ID'],axis=1)
y_train1=cv_train1['V86']
x_train2=cv_train2.drop(['V86','ID'],axis=1)
#y_train2=cv_train2['V86']
```

```python
grid_search.fit(x_train1,y_train1)
```

```python
grid_search.best_estimator_
```

```python
logcv=grid_search.best_estimator_
```

```python
#report(grid_search.cv_results_,5)
```

```python
logcv.fit(x_train1,y_train1)
```

```python
cutoffs=np.linspace(0.01,0.99,99)
```

```python
train1_score=logcv.predict_proba(x_train1)[:,1]
real=y_train1
KS_all=[]
F_beta_all=[]
mydata=[]
```

```python
for cutoff in cutoffs:
    
    predicted=(train1_score>cutoff).astype(int)

    TP=((predicted==1) & (real==1)).sum()
    TN=((predicted==0) & (real==0)).sum()
    FP=((predicted==1) & (real==0)).sum()
    FN=((predicted==0) & (real==1)).sum()
    
    P=TP+FN
    N=TN+FP
      
    KS=(TP/P)-(FP/N)
    Sn=TP/P
    Sp=TN/N
    precision=TP/(TP+FP)
    recall=Sn
    beta=2
    F_beta= (1+(beta**2))*((precision*recall)/(((beta**2)*precision)+recall))
  
  
    F5=(26*precision*recall)/((25*precision)+recall)
    F1=(1.01*precision*recall)/((.01*precision)+recall)
  
    M=(4*FP+FN)/(5*(P+N))
    
   # KS_all.append(KS)
    F_beta_all.append(F_beta)
    
    
#mydata = pd.concat([KS_all, F_beta_all], axis=1)
#pd.DataFrame(mydata, columns=['KS', 'F_beta']) 
```

```python
mycutoff=cutoffs[KS_all==max(KS_all)][0]
mycutoff
```

```python
mybetacutoff=cutoffs[F_beta_all==max(F_beta_all)][0]
mybetacutoff
```
